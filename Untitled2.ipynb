```python
import scipy.stats as stats

# Sample data
sample_mean = 50
std_dev = 5
confidence_level = 0.95
sample_size = 100  # Adjust sample size accordingly

# Calculate the margin of error
margin_of_error = round(stats.norm.ppf((1 + confidence_level) / 2),2) * std_dev / (sample_size ** 0.5)

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("95% Confidence Interval:", (lower_bound, upper_bound))
```

# 2.

```python
from scipy.stats import chisquare

# Observed frequencies of colors in the bag
observed_frequencies = [25, 18, 20, 12, 10, 15]  # Counts of blue, orange, green, yellow, red, brown

# Total count of observations
total_count = sum(observed_frequencies)

# Expected frequencies based on the expected distribution
expected_proportions = [0.2, 0.2, 0.2, 0.1, 0.1, 0.2]  # Expected proportions for blue, orange, green, yellow, red, brown
expected_frequencies = [expected_proportion * total_count for expected_proportion in expected_proportions]

# Perform the chi-square goodness of fit test
chi2_stat, p_value = chisquare(observed_frequencies, f_exp=expected_frequencies)

# Degrees of freedom
df = len(observed_frequencies) - 1

# Significance level
alpha = 0.05

# Determine critical value
critical_value = stats.chi2.ppf(1 - alpha, df)

# Print the results
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)
print("Degrees of freedom:", df)
print("Critical value:", critical_value)

# Compare test statistic and critical value
if chi2_stat > critical_value:
    print("Reject the null hypothesis: The observed distribution differs significantly from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution matches the expected distribution.")

```

```python
from scipy.stats import chi2_contingency

observed = [[20, 15],
            [10, 25],
            [15, 20]]

chi2_stat, p_val, _, _ = chi2_contingency(observed)

print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject null hypothesis: There is no significant association between smoking status and lung cancer diagnosis.")
```

```python
import scipy.stats as stats

sample_size = 500
smokers = 60

p_hat = smokers / sample_size
margin_of_error = 1.96 * ((p_hat * (1 - p_hat)) / sample_size) ** 0.5

confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)
print("95% Confidence Interval for proportion of smokers:", confidence_interval)
```

```python
import numpy as np
import matplotlib.pyplot as plt

df = 10
x = np.linspace(0, 30, 1000)
y = stats.chi2.pdf(x, df)

plt.plot(x, y, 'b-', lw=2)
plt.fill_between(x, y, where=(x >= 0) & (x <= 15), color='gray', alpha=0.5)
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density')
plt.title('Chi-square Distribution (df=10)')
plt.show()
```

```python
from scipy.stats import chisquare

observed = [45, 55]  # Observed frequencies of heads and tails respectively
expected = [50, 50]  # Expected frequencies for a fair coin

chi2_stat, p_val = chisquare(observed, f_exp=expected)

print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: The observed frequencies differ significantly from the expected frequencies of a fair coin.")
else:
    print("Fail to reject null hypothesis: The observed frequencies match the expected frequencies of a fair coin.")
```

```python
from scipy.stats import t

sample_mean = 72
sample_std = 10
n = 30
pop_mean = 70

t_stat = (sample_mean - pop_mean) / (sample_std / (n ** 0.5))
p_val = t.cdf(t_stat, df=n-1) * 2

print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in the population mean from 70.")
```
